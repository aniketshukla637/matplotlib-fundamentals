### 10 Simple Matplotlib Code Examples

These examples demonstrate fundamental plotting concepts in Matplotlib.

```python
# Example 1: Basic Line Plot
import matplotlib.pyplot as plt # Import the Matplotlib plotting library
import numpy as np # Import NumPy for numerical operations, especially for creating arrays

x = np.linspace(0, 10, 100) # Create an array of 100 evenly spaced numbers from 0 to 10
y = np.sin(x) # Calculate the sine of each number in x

plt.figure(figsize=(6, 4)) # Create a new figure with a specified size (width, height in inches)
plt.plot(x, y) # Plot the data: x values on the horizontal axis, y values on the vertical axis
plt.title('Simple Sine Wave') # Set the title of the plot
plt.xlabel('X-axis') # Set the label for the x-axis
plt.ylabel('Y-axis') # Set the label for the y-axis
plt.grid(True) # Add a grid to the plot for better readability
plt.show() # Display the plot
```

```python
# Example 2: Scatter Plot
import matplotlib.pyplot as plt # Import the Matplotlib plotting library
import numpy as np # Import NumPy for numerical operations

x = np.random.rand(50) # Generate 50 random numbers for the x-coordinates
y = np.random.rand(50) # Generate 50 random numbers for the y-coordinates
colors = np.random.rand(50) # Generate 50 random numbers for color mapping
area = (30 * np.random.rand(50))**2 # Generate 50 random numbers for marker size

plt.figure(figsize=(6, 4)) # Create a new figure
plt.scatter(x, y, s=area, c=colors, alpha=0.5) # Create a scatter plot with custom size (s), color (c), and transparency (alpha)
plt.title('Simple Scatter Plot') # Set the plot title
plt.xlabel('Random X') # Set the x-axis label
plt.ylabel('Random Y') # Set the y-axis label
plt.show() # Display the plot
```

```python
# Example 3: Bar Chart
import matplotlib.pyplot as plt # Import the Matplotlib plotting library

categories = ['A', 'B', 'C', 'D'] # Define categories for the x-axis
values = [23, 45, 56, 12] # Define numerical values for each category

plt.figure(figsize=(6, 4)) # Create a new figure
plt.bar(categories, values, color=['skyblue', 'lightcoral', 'lightgreen', 'gold']) # Create a bar chart with custom colors for each bar
plt.title('Simple Bar Chart') # Set the plot title
plt.xlabel('Category') # Set the x-axis label
plt.ylabel('Value') # Set the y-axis label
plt.show() # Display the plot
```

```python
# Example 4: Histogram
import matplotlib.pyplot as plt # Import the Matplotlib plotting library
import numpy as np # Import NumPy for numerical operations

data = np.random.randn(1000) # Generate 1000 random numbers from a standard normal distribution

plt.figure(figsize=(6, 4)) # Create a new figure
plt.hist(data, bins=30, edgecolor='black', alpha=0.7) # Create a histogram with 30 bins, black edges, and transparency
plt.title('Simple Histogram') # Set the plot title
plt.xlabel('Value') # Set the x-axis label
plt.ylabel('Frequency') # Set the y-axis label
plt.show() # Display the plot
```

```python
# Example 5: Pie Chart
import matplotlib.pyplot as plt # Import the Matplotlib plotting library

labels = ['Frogs', 'Hogs', 'Dogs', 'Logs'] # Labels for each slice of the pie
sizes = [15, 30, 45, 10] # Sizes (proportions) for each slice
explode = (0, 0.1, 0, 0)  # 'Explode' a slice by offsetting it from the center (e.g., the second slice by 10% of the radius)

plt.figure(figsize=(6, 6)) # Create a new figure, ensuring it's square for a circular pie
plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', # Create the pie chart; autopct formats percentage labels
        shadow=True, startangle=90) # Add a shadow and start the first slice at 90 degrees (top)
plt.axis('equal')  # Ensure the pie chart is drawn as a circle, not an ellipse.
plt.title('Simple Pie Chart') # Set the plot title
plt.show() # Display the plot
```

```python
# Example 6: Subplots (Multiple Plots in One Figure)
import matplotlib.pyplot as plt # Import the Matplotlib plotting library
import numpy as np # Import NumPy for numerical operations

x = np.linspace(0, 2 * np.pi, 400) # Create an array for the x-axis values
y_sin = np.sin(x) # Calculate sine values
y_cos = np.cos(x) # Calculate cosine values

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4)) # Create a figure and a set of subplots (1 row, 2 columns); returns figure and axes objects

ax1.plot(x, y_sin, color='blue') # Plot sine wave on the first subplot (ax1)
ax1.set_title('Sine Wave') # Set title for ax1
ax1.set_xlabel('Angle') # Set x-label for ax1
ax1.set_ylabel('Sine') # Set y-label for ax1
ax1.grid(True) # Add grid to ax1

ax2.plot(x, y_cos, color='red') # Plot cosine wave on the second subplot (ax2)
ax2.set_title('Cosine Wave') # Set title for ax2
ax2.set_xlabel('Angle') # Set x-label for ax2
ax2.set_ylabel('Cosine') # Set y-label for ax2
ax2.grid(True) # Add grid to ax2

plt.tight_layout() # Adjust subplot parameters for a tight layout, preventing labels from overlapping
plt.show() # Display the plot
```

```python
# Example 7: Adding a Legend
import matplotlib.pyplot as plt # Import the Matplotlib plotting library
import numpy as np # Import NumPy for numerical operations

x = np.linspace(0, 10, 100) # Create x values
y1 = np.sin(x) # Calculate y1 values (sine)
y2 = np.cos(x) # Calculate y2 values (cosine)

plt.figure(figsize=(6, 4)) # Create a new figure
plt.plot(x, y1, label='Sine Function', color='green') # Plot y1 with a label for the legend and a specific color
plt.plot(x, y2, label='Cosine Function', color='purple', linestyle='--') # Plot y2 with a label, color, and dashed line style
plt.title('Sine and Cosine with Legend') # Set the plot title
plt.xlabel('X-axis') # Set the x-axis label
plt.ylabel('Y-axis') # Set the y-axis label
plt.legend() # Display the legend, using the 'label' arguments from the plot calls
plt.grid(True) # Add a grid
plt.show() # Display the plot
```

```python
# Example 8: Box Plot
import matplotlib.pyplot as plt # Import the Matplotlib plotting library
import numpy as np # Import NumPy for numerical operations

# Generate some random data for three groups from normal distributions
data1 = np.random.normal(0, 1, 100) # Mean 0, Std Dev 1
data2 = np.random.normal(1, 2, 100) # Mean 1, Std Dev 2
data3 = np.random.normal(0.5, 0.5, 100) # Mean 0.5, Std Dev 0.5

data = [data1, data2, data3] # Combine the data into a list

plt.figure(figsize=(6, 4)) # Create a new figure
plt.boxplot(data, patch_artist=True, labels=['Group 1', 'Group 2', 'Group 3']) # Create a box plot; patch_artist fills boxes with color, labels for each box
plt.title('Simple Box Plot') # Set the plot title
plt.ylabel('Values') # Set the y-axis label
plt.grid(axis='y', linestyle='--', alpha=0.7) # Add a horizontal grid
plt.show() # Display the plot
```

```python
# Example 9: Customizing Line Styles and Markers
import matplotlib.pyplot as plt # Import the Matplotlib plotting library
import numpy as np # Import NumPy for numerical operations

x = np.linspace(0, 5, 20) # Create 20 evenly spaced x values from 0 to 5
y = x**2 # Calculate y values as x squared

plt.figure(figsize=(6, 4)) # Create a new figure
plt.plot(x, y, color='darkorange', linestyle=':', marker='o', markersize=5, label='y = x^2') # Plot with custom color, dotted line style, circular markers, marker size, and a label
plt.title('Customized Line Plot') # Set the plot title
plt.xlabel('X-axis') # Set the x-axis label
plt.ylabel('Y-axis') # Set the y-axis label
plt.legend() # Display the legend
plt.grid(True) # Add a grid
plt.show() # Display the plot
```

```python
# Example 10: Filling Area Under Curve
import matplotlib.pyplot as plt # Import the Matplotlib plotting library
import numpy as np # Import NumPy for numerical operations

x = np.linspace(0, 2 * np.pi, 100) # Create x values for a full cycle (0 to 2*pi)
y = np.cos(x) # Calculate y values (cosine)

plt.figure(figsize=(6, 4)) # Create a new figure
plt.plot(x, y, color='blue', alpha=0.7, label='Cosine Wave') # Plot the cosine wave with blue color and transparency
plt.fill_between(x, y, 0, where=(y > 0), color='skyblue', alpha=0.3, interpolate=True) # Fill area between the curve and y=0 where y is positive, with skyblue color
plt.fill_between(x, y, 0, where=(y < 0), color='lightcoral', alpha=0.3, interpolate=True) # Fill area between the curve and y=0 where y is negative, with lightcoral color
plt.title('Cosine Wave with Filled Area') # Set the plot title
plt.xlabel('Angle') # Set the x-axis label
plt.ylabel('Cosine') # Set the y-axis label
plt.legend() # Display the legend
plt.grid(True) # Add a grid
plt.show() # Display the plot
```
